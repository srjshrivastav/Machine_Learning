```python
weather_data=pd.read_csv('weather.csv')
weather_data_x=weather_data.values
```

```python
def set_class(i):
    return set(weather_data_x[:,i])
```

```python
def cal_positive(ch):
    count=0
    for i in range(weather_data_x.shape[0]):
        if weather_data_x[i,-1]==ch:
            count+=1
    return count
def cal_hpos(ch,j):
    count=0
    for i in range(weather_data_x.shape[0]):
        if (weather_data_x[i,j]==ch):
            if(weather_data_x[i,-1]=='Y'):
                count+=1
    return count
def cal_hneg(ch,j):
    count=0
    for i in range(weather_data_x.shape[0]):
        if (weather_data_x[i,j]==ch):
            if(weather_data_x[i,-1]=='N'):
                count+=1
    return count
    
def cal_neg(ch):
    count=0
    for i in range(weather_data_x.shape[0]):
        if weather_data_x[i,-1]==ch:
            count+=1
    return count
def count(ch,j):
    count=0
    for i in range(weather_data_x.shape[0]):
        if weather_data_x[i,j]==ch:
            count+=1
    return count
def H(ch,j):
    po=cal_hpos(ch,j)
    p1=cal_hneg(ch,j)
    c=count(ch,j)
    tem=0 if (p1==0) else p1/c*math.log(p1/c,2)
    tem2=0 if (po==0) else po/c*math.log(po/c,2)
    return -tem2-tem
```

```python
def cal_entropy():
    Po=cal_positive('Y')/weather_data_x.shape[0]
    p1=cal_neg('N')/weather_data_x.shape[0]
    return -Po*math.log(Po,2)-p1*math.log(p1,2)

cal_entropy()
def cal_sum(j):
    sum=0
    c=set_class(j)
    for i in c:
        sum=sum+(count(i,j)/weather_data_x.shape[0]*H(i,j))
    return sum
    
    
```

```python
IG=[]
def cal_IG(i):
    IG.append(round(cal_entropy()-cal_sum(i),3))
for i in range(1,weather_data_x.shape[1]-1):
    cal_IG(i)

print(IG)
print(weather_data.columns[IG.index(max(IG))+1])
```
